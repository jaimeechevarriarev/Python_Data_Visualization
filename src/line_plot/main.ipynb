# Line Plot

We will create a line plot of a sine function.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Data set 
x_lim = 2 * np.pi 
x_axis = np.arange(0, x_lim, 0.1)
sine = np.sin(x_axis)
```

```python
# View data 
print(x_lim)
print(x_axis)
print(sine)
```

```python
# Create plot
plt.title('Sine function from 0 to 2$\pi$')
plt.xlabel('Value of x')
plt.ylabel('sin(x)')

plt.hlines(0, 0, x_lim, alpha=0.2)
plt.xlim(0, x_lim)
plt.ylim(-1.25, 1.25)

plt.plot(x_axis, sine, color='blue', linewidth=1)
plt.show()
```
