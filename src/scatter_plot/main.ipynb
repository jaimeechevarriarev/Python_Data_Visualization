# Scatter Plot

We are going to generate random data to create a scatter plot using Matplotlib. The x-axis will represent the length of an input passed to the FakeSort algorithm (!) and be a range from 0.0 to 10.0 in 0.1 increments. The y-axis represents the time it took to sort the data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data
x_axis = np.arange(0, 10, 0.1)

y_axis = []

for x in x_axis: 
    y_axis.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
# Review values
print(x_axis)

print(y_axis)
```

```python
# Create plot
plt.title('Run time of FakeSort for Sample Input Size')
plt.xlabel('Length of Input')
plt.ylabel('Time to Sort (s)')

plt.scatter(x_axis, y_axis, marker='o', color='red')

plt.show()
```
